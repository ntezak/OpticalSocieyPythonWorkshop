### Hello World and Fibonacci Numbers

In the example below, we compute Fibonacci numbers in three ways: a pure-Python function, a Cython function with no type declarations, and a Cython function with type declarations.  Overall, Cython gives a 20x speedup in this case.

```python
import helloworld as hw
```

```python
hw.hi()
```

```python
def fib(n):
    a, b = 1, 1
    for i in range(n):
        c = a + b
        b = a
        a = c
    return a
```

Let's look at the first 40 Fibonacci numbers, below.  The Cython and Python functions give the same result, as expected.

```python
array([map(fib, range(40)), 
       map(hw.fib, range(40)),
       map(hw.fib_c, range(40))])
```

But the Cython implementation is 20x faster!

```python
%%timeit 
map(fib, range(40))
```

```python
%%timeit 
map(hw.fib, range(40))
```

```python
%%timeit
map(hw.fib_c, range(40))
```

### Inline Cython and Thermal Tunnelling

You can run Cython code in the iPython notebook using the `%%cython` cell magic.  To enable it, first call `%load_ext cythonmagic`, and place `%%cython` at the head of every Cython cell.

The code below, which simulates a thermally excited particle in a double-well potential, also demonstrates using `cdef` functions and the `cimport` statement.

```python
%load_ext cythonmagic
```

```python
%%cython

# Sloppy implementation.
# Calls NumPy rand() and exp(), Python overhead.
import numpy as np
from numpy.random import rand as pyRand

# U(x) is a cdef function.  This means that all its arguments are fixed-type, so it is
# guaranteed to run at C-speed.  Other Cython functions can call it without the Python
# function-call overhead.
cdef double U(double x):
    return (x-1)*(x-1) * (x+1)*(x+1)

cdef double stochastic_jump(double x, double T):
    cdef double x2 = x + 0.1*(pyRand() - 0.5)
    cdef double dU = U(x2) - U(x)
    cdef double p = pyRand()
    return x2 if (np.exp(-dU/T) > p) else x

def particle_in_box(double x0, int N, double T):
    cdef int i
    cdef double x = x0
    for i in range(N):
        x = stochastic_jump(x, T)
    return x
```

```python
%%timeit

particle_in_box(-1, 50000, 0.2)
```

```python
%%cython

# Correct implementation.
# Calls to C libraries, no Python overhead.
from libc.stdlib cimport rand, RAND_MAX
from libc.math cimport exp

cdef double U(double x):
    return (x-1)*(x-1) * (x+1)*(x+1)

cdef double rand_f():
    return (<double> rand())/RAND_MAX

cdef double stochastic_jump(double x, double T):
    cdef double x2 = x + 0.1*(rand_f() - 0.5)
    cdef double dU = U(x2) - U(x)
    cdef double p = rand_f()
    return x2 if (exp(-dU/T) > p) else x

def particle_in_box2(double x0, int N, double T):
    cdef int i
    cdef double x = x0
    for i in range(N):
        x = stochastic_jump(x, T)
    return x
```

```python
%%timeit

particle_in_box2(-1, 50000, 0.2)
```

```python
hist(array([particle_in_box2(-1, 50000, 0.2) for i in range(1000)]), 
     bins=40);
```

### Arrays in Cython

Cython supports fast access to arrays through objects called *typed memoryviews*.  Memoryviews are like C pointers, but they can have multiple indices and support bounds-checking to avoid segfaults.  But memoryviews lack some of the convenient features of NumPy arrays.  For instance, you can't add, subtract, or multiply them -- if so needed, you must first use `numpy.asarray()` to convert them to arrays, and then manipulate them.

```python
%%cython

import numpy as np

def cython_Ax(double [:, :] A, double [:] x):
    cdef double [:] Ax = np.zeros([A.shape[0]])
    cdef int i, j
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            Ax[i] += A[i, j] * x[j]
    return np.asarray(Ax)    
```

```python
A = rand(10, 10); x = rand(10); b = rand(10);
(cython_Ax(A, x), A.dot(x))
```

```python
%%cython

import numpy as np

def cython_Axpb(double [:, :] A, double [:] x, double [:] b):
    cdef double [:] Ax = np.zeros([A.shape[0]])
    cdef int i, j
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            Ax[i] += A[i, j] * x[j]
    Ax = Ax + b    # This doesn't work.
    return np.asarray(Ax)
```

```python
%%cython

import numpy as np

def cython_Axpb(double [:, :] A, double [:] x, double [:] b):
    cdef double [:] Ax = np.zeros([A.shape[0]])
    cdef int i, j
    for i in range(A.shape[0]):
        for j in range(A.shape[1]):
            Ax[i] += A[i, j] * x[j]
    Ax = np.asarray(Ax) + np.asarray(b)   # Needed to wrap memoryviews before adding!
    return np.asarray(Ax)
```

```python
(cython_Axpb(A, x, b), A.dot(x) + b)
```

### Demo 1: Logistic Map

The *logistic map* is the map

$$
    x \rightarrow r\,x(1-x)
$$

For $r \gtrsim 3.6$, this map becomes chaotic.  To see this, we would like to plot the probability distribution of $x$ as a function of $r$.  We can do that with the following slow Python code:

```python
# Slow Python code.
def logistic_map(x_n, r):
    return r * x_n * (1 - x_n)

def draw_logistic_map(x_0, n_r, n_x, N, N0=20, r1=2.5, r2=4.):
    out = zeros([n_r, n_x], dtype=int)
    rList = linspace(r1, r2, n_r)
    for (r_ind, r) in enumerate(rList):
        x = x_0
        for i in range(N0):
            x = logistic_map(x, r)
        for i in range(N):
            x = logistic_map(x, r)
            x_ind = int(x * n_x)
            out[r_ind, x_ind] += 1
    return out
```

```python
%%timeit
draw_logistic_map(0.1, 300, 300, 1000, 400)
```

```python
lm = draw_logistic_map(0.1, 300, 300, 1000, 400)
figure(figsize=(8, 6))
imshow(lm.transpose()+1, origin='lower',
       extent=[2.5, 4, 0, 1], aspect='auto',
       norm=matplotlib.colors.LogNorm());
xlabel(r"$r$"); ylabel(r"$x$"); title("Logistic Map");
```

```python
%%cython
# Just wrapped everything in Cython.  Didn't set any type declarations.

import numpy as np

def logistic_map_c(x_n, r):
    return r * x_n * (1 - x_n)

def draw_logistic_map_c(x_0, n_r, n_x, N, N0=20):
    out = np.zeros([n_r, n_x], dtype=np.int32)
    rList = np.linspace(2.5, 4, n_r)
    for (r_ind, r) in enumerate(rList):
        x = x_0
        for i in range(N0):
            x = logistic_map_c(x, r)
        for i in range(N):
            x = logistic_map_c(x, r)
            x_ind = int(x * n_x)
            out[r_ind, x_ind] += 1
    return out
```

```python
%%timeit
draw_logistic_map_c(0.1, 300, 300, 1000, 400)
```

Well, that wasn't any better.

```python
%%cython
# Set a few type declarations.  However, no memoryviews yet.

import numpy as np

cdef double logistic_map_c(double x_n, double r):
    return r * x_n * (1 - x_n)

def draw_logistic_map_c(double x_0, int n_r, int n_x, int N, int N0=20):
    cdef int r_ind, i, x_ind
    cdef double r, x
    out = np.zeros([n_r, n_x], dtype=np.int32)
    rList = np.linspace(2.5, 4, n_r)
    for r_ind in range(n_r):
        r = rList[r_ind]
        x = x_0
        for i in range(N0):
            x = logistic_map_c(x, r)
        for i in range(N):
            x = logistic_map_c(x, r)
            x_ind = <int> (x * n_x)
            out[r_ind, x_ind] += 1
    return out
```

```python
%%timeit
draw_logistic_map_c(0.1, 300, 300, 1000, 400)
```

```python
%%cython
# Fully optimized: arrays are accessed as memoryviews, no bounds-checking on indices.

cimport cython
import numpy as np

cdef double logistic_map_c(double x_n, double r) nogil:
    return r * x_n * (1 - x_n)

@cython.boundscheck(False)
@cython.wraparound(False)
def draw_logistic_map_c(double x_0, int n_r, int n_x, int N, int N0=20,
                        double r1=2.5, double r2=4.):
    cdef int r_ind, i, x_ind
    cdef double r, x
    cdef int [:, ::1] out = np.zeros([n_r, n_x], dtype=np.int32)
    cdef double [::1] rList = np.linspace(r1, r2, n_r)
    for r_ind in range(n_r):
        r = rList[r_ind]
        x = x_0
        for i in range(N0):
            x = logistic_map_c(x, r)
        for i in range(N):
            x = logistic_map_c(x, r)
            x_ind = <int> (x * n_x)
            out[r_ind, x_ind] += 1
    return np.asarray(out)
```

```python
%%timeit
draw_logistic_map_c(0.1, 300, 300, 1000, 400)
```

And that is fast enough to make this into an interactive widget!

```python
lm = draw_logistic_map_c(0.1, 500, 500, 10000, 400, 3.4, 4.0)
figure(figsize=(10, 8))
imshow(lm.transpose()+1, origin='lower',
       extent=[3.53, 3.6, 0, 1], aspect='auto',
       norm=matplotlib.colors.LogNorm());
xlabel(r"$r$"); ylabel(r"$x$"); title("Logistic Map");
```

### Exercise: Julia Set

The following code computes the Julia set, but it is very slow.  Optimize it using Cython to get a 500x speedup in execution time.

```python
import numpy as np
from time import time

def kernel(z, c, lim, cutoff=1e6):
    ''' Computes the number, `n`, of iterations necessary such that 
    |z_n| > `lim`, where `z_n = z_{n-1}**2 + c`.
    '''
    count = 0
    while abs(z) < lim and count < cutoff:
        z = z * z + c
        count += 1
    return count


def compute_julia(c, N, bound=2, lim=1000., kernel=kernel):
    ''' Pure Python calculation of the Julia set for a given `c`.  No NumPy
    array operations are used.
    '''
    julia = np.empty((N, N), dtype=np.uint32)
    grid_x = np.linspace(-bound, bound, N)
    grid_y = grid_x * 1j
    c = complex(c)
    t0 = time()
    for i, x in enumerate(grid_x):
        for j, y in enumerate(grid_y):
            julia[i,j] = kernel(x+y, c, lim)
    return julia, time() - t0
```

```python
import pylab as pl

def plot_julia(kwargs, compute_julia):
    ''' Given parameters dict in `kwargs` and a function to compute the Julia
    set (`compute_julia`), plots the resulting Julia set with appropriately
    labeled axes.
    '''
    kwargs = kwargs.copy()

    def _plotter(kwargs):
        bound = kwargs['bound']
        julia, time = compute_julia(**kwargs)
        print "execution time (s):", time
        julia = np.log(julia)
        pl.imshow(julia, 
                  interpolation='nearest',
                  extent=(-bound, bound)*2)
        pl.colorbar()
        title = r"Julia set for $C={0.real:5.3f}+{0.imag:5.3f}i$ $[{1}\times{1}]$"
        pl.title(title.format(kwargs['c'], kwargs['N']))
        pl.xlabel("$Re(z)$")
        pl.ylabel("$Im(z)$")

    pl.figure(figsize=(14, 12))

    cvals = [0.285+0.01j, -0.1+0.651j, -0.4+0.6j, -0.8+0.156j]
    subplots = ['221',    '222',       '223',     '224'      ]

    for c, sp in zip(cvals, subplots):
        kwargs.update(c=c)
        pl.subplot(sp)
        _plotter(kwargs)

    pl.show()
```

```python
kwargs = dict(N=100, bound=1.5)
plot_julia(kwargs, compute_julia)
```
